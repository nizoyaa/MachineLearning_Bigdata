```python
# Kolom Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen)
df['Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen)'] = df['Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen)'].str.replace(',', '.')

# Kolom Rata-rata Lama Sekolah Penduduk 15+ (Tahun)
df['Rata-rata Lama Sekolah Penduduk 15+ (Tahun)'] = df['Rata-rata Lama Sekolah Penduduk 15+ (Tahun)'].str.replace(',', '.')

# Kolom Indeks Pembangunan Manusia
df['Indeks Pembangunan Manusia'] = df['Indeks Pembangunan Manusia'].str.replace(',', '.')

# Kolom Persentase rumah tangga yang memiliki akses terhadap sanitasi layak
df['Persentase rumah tangga yang memiliki akses terhadap sanitasi layak'] = df['Persentase rumah tangga yang memiliki akses terhadap sanitasi layak'].str.replace(',', '.')

# Kolom Persentase rumah tangga yang memiliki akses terhadap air minum layak
df['Persentase rumah tangga yang memiliki akses terhadap air minum layak'] = df['Persentase rumah tangga yang memiliki akses terhadap air minum layak'].str.replace(',', '.')

# Kolom Tingkat Pengangguran Terbuka
df['Tingkat Pengangguran Terbuka'] = df['Tingkat Pengangguran Terbuka'].str.replace(',', '.')

# Kolom Tingkat Partisipasi Angkatan Kerja
df['Tingkat Partisipasi Angkatan Kerja'] = df['Tingkat Partisipasi Angkatan Kerja'].str.replace(',', '.')

```

```python
# Membuat variabel untuk menyimpan nilai awal ID kab/kota untuk setiap provinsi
start_id_per_provinsi = {
    'ACEH': 101,
    'SUMATERA UTARA': 201,
    'SUMATERA BARAT': 301,
    'RIAU':401,
    'JAMBI':501,
 'SUMATERA SELATAN': 601,
'BENGKULU':701,
'LAMPUNG':801,
'KEP. BANGKA BELITUNG':901,
    'KEPULAUAN RIAU': 1001,
        'DKI JAKARTA': 1101,
        'JAWA BARAT': 1201,
        'JAWA TENGAH': 1301,
'D I YOGYAKARTA':1401,
    'JAWA TIMUR' :1501,
    'BANTEN':1601,
    'BALI':1701,
    'NUSA TENGGARA BARAT':1801,
'NUSA TENGGARA TIMUR':1901,
    'KALIMANTAN BARAT':2001,
    'KALIMANTAN TENGAH':2101,
     'KALIMANTAN SELATAN':2201,
    'KALIMANTAN TIMUR' :2301,
    'KALIMANTAN UTARA':2401,
    'SULAWESI UTARA':2501,
    'SULAWESI TENGAH' :2601,
    'SULAWESI SELATAN':2701,
    'SULAWESI TENGGARA':2801,
    'GORONTALO':2901,
        'SULAWESI BARAT':3001,
    'MALUKU' :3101,
    'MALUKU UTARA' :3201,
    'PAPUA BARAT':3201,
 'PAPUA':3401

}

# Membuat variabel untuk menyimpan nilai ID kab/kota
id_kab_kota = {}

# Mendefinisikan fungsi untuk menetapkan ID kab/kota berdasarkan provinsi
def assign_kab_id(row):
    global id_kab_kota
    provinsi = row['Provinsi']
    if provinsi not in id_kab_kota:
        id_kab_kota[provinsi] = start_id_per_provinsi.get(provinsi, 1)
    else:
        id_kab_kota[provinsi] += 1
    return id_kab_kota[provinsi]

# Menetapkan nilai ID kab/kota menggunakan fungsi yang telah didefinisikan
df['ID Kab/Kota'] = df.apply(assign_kab_id, axis=1)

# Menampilkan DataFrame hasil
print(df[['Provinsi', 'ID Kab/Kota']])

```

```python
# Menyimpan posisi kolom 'ID Provinsi'
pos_id_provinsi = df.columns.get_loc('ID_Provinsi')

# Memindahkan kolom 'ID Kab/Kota' setelah kolom 'ID Provinsi'
df.insert(pos_id_provinsi + 1, 'ID_Kab/Kota', df.pop('ID Kab/Kota'))

# Menampilkan DataFrame hasil
print(df)
```
